```python
# ================================================
# Artify an AVIF image via Neural Style Transfer (VGG19, ReLU inplace fix)
# ================================================
from pathlib import Path
import os
import numpy as np
from PIL import Image, UnidentifiedImageError
# Try to enable AVIF support if installed
for _plugin in ("pillow_avif_plugin", "pillow_avif"):
    try:
        __import__(_plugin)  # registers AVIF handler with Pillow if present
        break
    except Exception:
        pass

import torch
import torch.nn as nn
import torch.optim as optim
import torchvision.models as models
import torchvision.transforms as T

# ---------- USER PATHS ----------
CONTENT_PATH = r"C:\Users\sagni\Downloads\Artify AI\flat-lay-acrylic-paint-background_23-2149127621.avif"
STYLE_SEARCH_DIR = Path(r"C:\Users\sagni\Downloads\Artify AI\archive")   # where to auto-pick a style image
OUT_DIR = Path(r"C:\Users\sagni\Downloads\Artify AI\outputs")
OUT_DIR.mkdir(parents=True, exist_ok=True)

# ---------- NST Settings (tweak as you like) ----------
MAX_SIZE       = 768       # longest side (keep modest on CPU)
STEPS          = 350       # ~300700
CONTENT_WEIGHT = 1e5
STYLE_WEIGHT   = 1e6
TV_WEIGHT      = 1e-5
LR             = 0.02
STYLE_BLEND    = None      # e.g., [0.7, 0.3] if you set two style images
PRESERVE_COLOR = False     # set True to approximately keep content colors (needs opencv-python)

IMG_EXTS = {".jpg", ".jpeg", ".png", ".bmp", ".tif", ".tiff", ".webp", ".avif"}

# ---------- Utilities ----------
def load_image(path, max_size=None):
    img = Image.open(path).convert("RGB")
    if max_size is not None:
        w, h = img.size
        scale = max_size / max(w, h)
        if scale < 1.0:
            img = img.resize((int(w * scale), int(h * scale)), Image.LANCZOS)
    return img

def pil_to_tensor(img_pil, device):
    tfm = T.Compose([
        T.ToTensor(),
        T.Normalize(mean=[0.485, 0.456, 0.406],
                    std=[0.229, 0.224, 0.225])
    ])
    return tfm(img_pil).unsqueeze(0).to(device)

def tensor_to_pil(tensor):
    tensor = tensor.detach().cpu().clone().squeeze(0)
    unnorm = T.Normalize(mean=[-m/s for m, s in zip([0.485,0.456,0.406],[0.229,0.224,0.225])],
                         std=[1/s for s in [0.229,0.224,0.225]])
    img = unnorm(tensor).clamp(0, 1)
    return T.ToPILImage()(img)

def match_color_simple(content_pil, stylized_pil):
    """Approx color preservation via LAB stats (requires opencv-python)."""
    try:
        import cv2
        c = cv2.cvtColor(np.array(content_pil), cv2.COLOR_RGB2LAB).astype(np.float32)
        s = cv2.cvtColor(np.array(stylized_pil), cv2.COLOR_RGB2LAB).astype(np.float32)
        c_mean, c_std = c.mean(axis=(0,1)), c.std(axis=(0,1)) + 1e-6
        s_mean, s_std = s.mean(axis=(0,1)), s.std(axis=(0,1)) + 1e-6
        matched = (s - s_mean) / s_std * c_std + c_mean
        matched = np.clip(matched, 0, 255).astype(np.uint8)
        out = cv2.cvtColor(matched, cv2.COLOR_LAB2RGB)
        return Image.fromarray(out)
    except Exception:
        return stylized_pil

def find_style_images(search_dir: Path, exclude_path: str, max_styles=1):
    if not search_dir.exists():
        return []
    files = []
    for p in search_dir.rglob("*"):
        if p.is_file() and p.suffix.lower() in IMG_EXTS and str(p) != exclude_path:
            files.append(p)
    # simple heuristic: pick first N (you can sort by size/date if you want)
    return [str(p) for p in files[:max_styles]]

# ---------- VGG Feature Extractor (ReLU inplace=False) ----------
class VGGFeatures(nn.Module):
    def __init__(self, layers_content, layers_style):
        super().__init__()
        vgg = models.vgg19(weights=models.VGG19_Weights.IMAGENET1K_V1).features
        # disable in-place ReLUs to avoid autograd errors
        for i, m in enumerate(vgg):
            if isinstance(m, nn.ReLU):
                vgg[i] = nn.ReLU(inplace=False)
        for p in vgg.parameters():
            p.requires_grad_(False)
        self.vgg = vgg.eval()
        self.layers_content = set(layers_content)
        self.layers_style = set(layers_style)

    def forward(self, x):
        content_feats, style_grams = {}, {}
        for name, layer in self.vgg._modules.items():
            x = layer(x)
            if name in self.layers_content:
                content_feats[name] = x
            if name in self.layers_style:
                b, c, h, w = x.shape
                feat = x.view(b, c, h*w)
                gram = torch.bmm(feat, feat.transpose(1,2)) / (c*h*w)
                style_grams[name] = gram
        return content_feats, style_grams

# ---------- NST Core ----------
def stylize(content_path, style_paths, blend_weights=None, size=768, steps=400,
            content_weight=1e5, style_weight=1e6, tv_weight=1e-5, lr=0.02, device="cpu"):
    content_pil = load_image(content_path, max_size=size)
    content_t = pil_to_tensor(content_pil, device)

    style_tensors = [pil_to_tensor(load_image(p, max_size=size), device) for p in style_paths]
    if blend_weights is None or len(blend_weights) != len(style_tensors):
        blend_weights = [1.0/len(style_tensors)] * len(style_tensors)
    S = sum(blend_weights)
    blend_weights = [w / S for w in blend_weights]

    layers_content = ["21"]                    # relu4_2
    layers_style   = ["0","5","10","19","28"]  # relu1_1..relu5_1
    extractor = VGGFeatures(layers_content, layers_style).to(device)

    with torch.no_grad():
        cF, _ = extractor(content_t)
        target_style = {k: 0 for k in layers_style}
        for s_t, w in zip(style_tensors, blend_weights):
            _, sF = extractor(s_t)
            for k in layers_style:
                target_style[k] = target_style[k] + w * sF[k]

    x = content_t.clone().requires_grad_(True)
    opt = optim.Adam([x], lr=lr)

    def tv_loss(x):
        a = torch.mean(torch.abs(x[:, :, :, :-1] - x[:, :, :, 1:]))
        b = torch.mean(torch.abs(x[:, :, :-1, :] - x[:, :, 1:, :]))
        return a + b

    for i in range(1, steps+1):
        opt.zero_grad()
        cFx, sFx = extractor(x)

        c_loss = sum(torch.mean((cFx[k] - cF[k])**2) for k in layers_content)
        s_loss = sum(torch.mean((sFx[k] - target_style[k])**2) for k in layers_style)
        t_loss = tv_loss(x)

        loss = content_weight * c_loss + style_weight * s_loss + tv_weight * t_loss
        loss.backward()
        opt.step()

        if i % max(1, steps//10) == 0 or i == 1:
            print(f"[{i:4d}/{steps}] total={float(loss):.4e}  "
                  f"C={float(c_loss):.4e} S={float(s_loss):.4e} TV={float(t_loss):.4e}")

    out_pil = tensor_to_pil(x)
    return content_pil, out_pil

# ---------- Run ----------
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"[INFO] Device: {device}")

# Pick a style image automatically (if available), else fall back to content
style_candidates = find_style_images(STYLE_SEARCH_DIR, CONTENT_PATH, max_styles=1)
if not style_candidates:
    print("[WARN] No style images found in archive; using the content image itself as the style (effect will be minimal).")
    STYLE_PATHS = [CONTENT_PATH]
else:
    STYLE_PATHS = style_candidates
print("[INFO] Style image(s):", STYLE_PATHS)

# Execute NST
content_pil, stylized_pil = stylize(
    content_path=CONTENT_PATH,
    style_paths=STYLE_PATHS,
    blend_weights=STYLE_BLEND,
    size=MAX_SIZE,
    steps=STEPS,
    content_weight=CONTENT_WEIGHT,
    style_weight=STYLE_WEIGHT,
    tv_weight=TV_WEIGHT,
    lr=LR,
    device=device
)

if PRESERVE_COLOR:
    stylized_pil = match_color_simple(content_pil, stylized_pil)

# Save result
c_name = Path(CONTENT_PATH).stem
if len(STYLE_PATHS) == 1:
    s_name = Path(STYLE_PATHS[0]).stem
else:
    s_name = "blend_" + "_".join([Path(p).stem for p in STYLE_PATHS])[:80]

out_img = OUT_DIR / f"artified_{c_name}_with_{s_name}.png"
stylized_pil.save(out_img)
print(f"[SAVE] Stylized -> {out_img}")

# (Optional) inline side-by-side preview if running in Jupyter
try:
    from IPython.display import display
    w1, h1 = content_pil.size
    w2, h2 = stylized_pil.size
    H = max(h1, h2)
    c_res = content_pil.resize((int(w1*H/h1), H)) if h1 != H else content_pil
    s_res = stylized_pil.resize((int(w2*H/h2), H)) if h2 != H else stylized_pil
    side = Image.new("RGB", (c_res.width + s_res.width, H), (255, 255, 255))
    side.paste(c_res, (0, 0))
    side.paste(s_res, (c_res.width, 0))
    display(side)
except Exception:
    pass
```
